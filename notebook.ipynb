### Structure of the FastAPI Application

```python
# Import FastAPI class
from fastapi import FastAPI

# Create an instance of FastAPI
app = FastAPI()

# Root endpoint
@app.get("/")
def home():
    # Returns a JSON response
    return {"message": "Hello World"}

# Second endpoint
@app.get("/about")
def about():
    return {"info": "This is a beginner FastAPI project"}
```
